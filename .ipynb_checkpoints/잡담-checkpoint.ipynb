```python
def mergeAndSortLists(list_1,list_2):
    list_3=list_1 +list_2
    list_3.sort()
    for item_1 in list_3:
        if list_3.count(item_1) ==1:
            continue
        else:
            while list_3.count(item_1)>1:
                list_3.remove(item_1)
    return list_3
```

```python
import turtle as t
def drawShape(alpha,beta):
    
    for due in range(0,beta):
        t.forward(alpha)
        t.left(360/beta)
line_len= int(input('insert the line length'))
shape=int(input('insert the shape of the drawing'))
drawShape(line_len,shape)
```

```python
import random

gamelist=['rock','scissor','paper']
gameresult=['tie','you lose','you win']
gamehistory=[]


def generatecomputerchoice():
    random_1=random.randint(1,3)
    if random_1==1:
        return 'rock'
    elif random_1==2:
        return 'scissor'
    else:
        return 'paper'
    
def verifyuserchoice(input_user):
    if input_user in gamelist:
        return True
    else:
        return False
    
def makedecision(input_user,com_input):
    if input_user =='rock':
        if com_input=='rock':
            b=0
        elif com_input=='scissor':
            b=2
        else:
            b=1
    elif input_user=='scissor':
        if com_input=='scissor':
            b=0
        elif com_input=='paper':
            b=2
        else:
            b=1 
    else:
        if com_input=='paper':
            b=0
        elif com_input=='rock':
            b=2
        else:
            b=1
    gamehistory.append(gameresult[b])
    return gameresult[b]

def printgamehistory():
    print('gamehistory :',gamehistory)
    

def allgame():
    user_input=input('enter yout action (rock,scissor,paper,quit)')
    while True:
        if user_input=='quit':
            print('program quit')
            printgamehistory()
            break
        else:
            if verifyuserchoice(user_input):
                com_input=generatecomputerchoice()
                dialogue='your choice is {0} vs compuuter choice is {1}'.format(user_input,com_input)
                print(dialogue,'>>>>',makedecision(user_input,com_input))
                
            else:
                print('wrong input')
            user_input=input('enter yout action (rock,scissor,paper,quit)')
            
            
allgame()
```

```python
Input_num=int(input('enter the num'))
for firstraw_item in range(1,Input_num+1):
    print('\t',firstraw_item,end='')
print()
for i in range(1,Input_num+1):
    print(i,end='')
    for j in range(1,Input_num+1):
        if i<j:
            print('\t','+',end='')
        elif i==j:
            print('\t','-',end='')
        elif (i%2==0) and (j%2==0):
            print('\t','*',end='')
        else:
            print('\t',j+i,end='')
    print()
        
        
```
