Answer for Question.1

```python
# Start of Answer
def CalcFactorial(input_1):
    if type(input_1)!=int:
        return -1
    elif input_1<=0:
        return -2
    else:
        result=1
        for i in range(1,input_1+1):
            result*=i 
        return result
# End of Answer

# Below is Execution Example

print(CalcFactorial(5))
```

```python
caseTypes = ["scissor", "rock", "paper"]

# Start of Answer
import random

gamelist=['rock','scissor','paper']
gameresult=['Tie','Computer','User']

def GenerateRandomCaseForComputer():
    random_1=random.randint(1,3)
    if random_1==1:
        return 'rock'
    elif random_1==2:
        return 'scissor'
    else:
        return 'paper'
    
def MakeDecision(input_user,com_input):
    if (input_user in gamelist) and (com_input in gamelist):
        if input_user =='rock':
            if com_input=='rock':
                b=0
            elif com_input=='scissor':
                b=2
            else:
                b=1
        elif input_user=='scissor':
            if com_input=='scissor':
                b=0
            elif com_input=='paper':
                b=2
            else:
                b=1 
        else:
            if com_input=='paper':
                b=0
            elif com_input=='rock':
                b=2
            else:
                b=1
        return gameresult[b]
    else:
        return -1
# End of Answer

# Below is Execution Example

userInput = 0
while True:
    userInput = int(input("\nMenu: \n[0] scissor, \n[1] rock, \n[2] paper, \n[3] quit \n\nSelect: "))
    if(userInput != 3):
        valueUser = caseTypes[userInput]
        valueComputer = GenerateRandomCaseForComputer()
        valueDecision = MakeDecision(valueUser, valueComputer)
        print("\nResult: User [{0}] vs Computer [{1}] -> Winner is {2}".format(valueUser, valueComputer, valueDecision))
    else:
        break
```
